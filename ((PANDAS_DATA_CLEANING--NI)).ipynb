```python
def fun1(x):
    if 'Touchscreen' in x:
        return 1
    else:
        return 0
df['Touch_screen']=df["ScreenResolution"].apply(fun1)
```

```python
def fun2(x):
    if "IPS" in x:
        return 1
    else:
        return 0
df['IPS']=df["ScreenResolution"].apply(fun2)
```

```python
#devide this IPS Panel Retina Display 2560x1600 data into two parts 1 is IPS Panel Retina Display 2560 2nd  1600
new=df['ScreenResolution'].str.split('x',n=1,expand=True)
```

```python
#add this data into data frame, we cteated 2 new columns
df['x_resolution']=new[0]
df['y_resolution']=new[1]
df
```

```python
# in x_resolution column we have data IPS Panel Full HD / Touchscreen 1920 like this we keep only integer
import re
def fun3(x):
    return re.findall('\d+',x)
```

```python
df['x_resolution']=df['x_resolution'].apply(fun3) #here we get data in the list of form [xxx]like this we want only xxxx
```

```python
def fun4(x):
    return x[0]
df['x_resolution']=df['x_resolution'].apply(fun4)
```
